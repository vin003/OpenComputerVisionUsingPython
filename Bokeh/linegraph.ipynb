<strong>Drawing the Graph  using Bokeh  module </strong> ; 



```python
from bokeh.plotting import figure
from bokeh.io import output_file,show 

x1=[1,2,3,4,5]
y=list(map(lambda x:x**2,x1))
print(y)
```

```python
f=figure()  ## to create the differenttyp of figures create an object 
output_file("line.html")
f.line(x1,y)
show(f)
```
